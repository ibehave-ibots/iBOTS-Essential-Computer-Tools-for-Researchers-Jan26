## Demo: Interactive Widgets

We can also add interactive elements to a notebook, for example using the `ipywidgets` package. Run the cell below to create a widget that plots a sinewave with variable frequency and amplitude.
Note that this only works if you installed the packages `numpy`, `matplotlib` and `ipywidgets` in the previous exercises.

````python
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt

# Create interactive plot with slider
def plot_sine(frequency=1.0, amplitude=1.0):
    x = np.linspace(0, 4*np.pi, 200)
    y = amplitude * np.sin(frequency * x)
    
    plt.figure(figsize=(10, 4))
    plt.plot(x, y)
    plt.ylim(-5, 5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'y = {amplitude:.1f} * sin({frequency:.1f} * x)')
    plt.grid(True)
    plt.show()

# Create interactive sliders
widgets.interactive(plot_sine, 
                   frequency=(0.5, 5.0, 0.1), 
                   amplitude=(0.5, 5.0, 0.1))
````
